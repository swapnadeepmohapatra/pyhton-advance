```python
import pandas as pd
import bs4
import requests
import csv
```

```python
res = requests.get(
    'https://www.flipkart.com/search?q=realme&otracker=search&otracker1=search&marketplace=FLIPKART&as-show=on&as=off')
soup = bs4.BeautifulSoup(res.text, 'lxml')
```

```python
soup
```

```python
phone = (soup.select('._2cLu-l'))
price = (soup.select('._1vC4OE'))
```
